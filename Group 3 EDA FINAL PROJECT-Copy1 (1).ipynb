```python
#create a function for customer group from customer since to determine old and new customers
def cus_group(x):
    if x <= 2006:
        return "Old Customer(From 1986-2006)"
    else:
        return "New Customer(From 2006-2016)"
```

1.	Gender count & Percentage of order by gender

Recommendation
Gender Preferences: Analyze the purchasing patterns and preferences of male and female customers to identify any significant differences. This analysis can help the company to understand which product categories or specific items are more popular among each gender.

New Customer Acquisition: The higher count of new customers ordering products compared to old customers indicates that Sterling E-commerce is successful in attracting new users to the platform. This is a positive sign as it shows a growing customer base.
However, 
2b.	Distributions of customers from 1986 to 2016
Retention and Loyalty: The right-skewed distribution suggests that there are fewer old customers ordering products compared to new customers. 
Recommendations for 2a and 2b
i.	Customer Retention Strategies: Implement strategies to retain existing customers and encourage repeat purchases. This can include loyalty programs, personalized offers, exclusive discounts, and excellent customer service. Building strong relationships with existing customers can lead to long-term loyalty and higher customer satisfaction.
ii.	New Customer Onboarding: Enhance the onboarding process for new customers to ensure a smooth and positive experience. Provide clear instructions, user-friendly interfaces, and proactive support to help new customers navigate the platform and make their first purchases.
3.	Count of Order by Quarter
Count of Order by Quarter showed significant increase at 4th quarter followed by 2nd quarter, followed by 1st quarter, with the 3rd quarter having the least count of order.
Insight
Seasonal Variation: The significant increase in orders during the 4th quarter suggests a seasonal trend, likely influenced by holiday shopping, such as Christmas and end-of-year sales. This indicates that customers are more active and make more purchases during the holiday season. The 2nd quarter also shows a notable increase, which could be influenced by factors such as seasonal promotions or events such as summer vacations.
Recommendations
i.	Seasonal Promotions: Capitalize on the increased demand during the 4th quarter and 2nd quarter by offering targeted seasonal promotions, discounts, and special deals. Create marketing campaigns that align with holidays, events, and customer preferences during these periods to maximize sales and customer engagement.
ii.	Off-Season Strategies: Develop strategies to boost sales and customer engagement during the 3rd quarter when order counts are relatively lower. This could involve implementing targeted marketing campaigns, introducing exclusive offers, or collaborating with partners to create special promotions to stimulate demand during this period.

4.	Count of order by State & Count of order by region
Count of order by state showed that; Texas is the highest selling state followed by, California, New York, and Pennsylvania respectively. This indicates that these states have a significant customer base and demand for products offered by Sterling E-commerce. And according to our research on google, these states are all populous states with large markets.
Furthermore, the Count of order by region; indicates “regional sales distribution” because Texas is in the southern region of the USA, which is why the southern region has the highest order as seen in the chart followed by the Midwest region where California; the second highest selling state is located.
Recommendations
Sterling E-Commerce should Build upon the success in Texas, California, New York, and Pennsylvania by implementing targeted marketing campaigns specific to these states while also consider expanding the customer base by targeting regions with untapped potential.

6.	Count of Payment Method & Revenue by Payment method
The most preferred payment method among customers is COD (Cash on Delivery), indicating that a significant portion of customers prefers the convenience of paying in cash upon delivery. Following COD, the second most preferred payment method is Easypay.
Although COD is the most preferred payment method, the highest revenue is generated from Easypay. This suggests that even though customers may choose COD for payment, transactions made through Easypay tend to have higher order values on average. Additionally, Bank Alfalah and Easypay Voucher are also generating significant revenue.
Recommendations
Promote Easypay: Since Easypay generates higher revenue despite being the second most preferred payment method, it would be beneficial for Sterling E-commerce to promote and incentivize customers to choose Easypay. This can be done by offering discounts, exclusive deals, or additional benefits for customers using Easypay.
Analyze Bank Alfalah and Easypay Voucher Usage Pattern to Identify the customer segments or marketing channels that drive transactions through these payment methods and strategize marketing efforts accordingly to further increase revenue from them.
7.	Count of Payment method by region 
The south and the Midwest has the highest “count of payment method by region” which is the same trend as the “count of order by region”. Indicating that customers in the South and Midwest regions of the USA have a similar preference for payment methods as they do for placing orders. This alignment between the count of payment methods and the count of orders by region suggests a correlation between regional shopping behaviour and payment preferences.
The fact that the South and Midwest regions have the highest count of both “payment methods” and “orders” indicates that customers in these regions are actively engaged in online shopping.
Recommendations
Therefore, it is essential for Sterling E-commerce to optimize the availability and usability of the preferred payment methods in these regions. Ensure that the payment infrastructure is robust, reliable, and convenient for customers using their preferred payment methods.
8.	Count of Payment method by cus_group
The analysis of the count of payment methods by customer group indicates that there is a higher trend in the "New Customer (From 2006-2016)" category compared to the "Old Customer (From 1986-2006)" category. This suggests that new customers, who have joined Sterling E-commerce more recently, are utilizing a wider variety of payment methods compared to the older customers. Indicating the willingness of new customers to explore and adopt different payment options.
Recommendations
Education and Communication: Provide clear and comprehensive information about the available payment methods to both new and old customers. Highlight the advantages, security features, and ease of use for each payment option. This can help bridge the knowledge gap and encourage older customers to explore and utilize a wider range of payment methods.
Personalization and Customization: Leverage customer data and insights to personalize the payment experience.
Seamless Integration: Ensure a seamless and user-friendly payment experience across all platforms and devices. Optimize the checkout process, streamline payment flows, and minimize friction points to make it convenient for customers to complete their transactions using their preferred payment methods.
